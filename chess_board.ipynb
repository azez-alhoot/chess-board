# Chess Board Lab-1

```python
import numpy as np
import matplotlib.pyplot as plt

class ChessBoard:
    blue = []
    red = []
    def __init__(self):
        self.grid = np.zeros([8,8,3])
        for i in range(8):
            for j in range(8):
                if (i%2 == 0 and j%2 == 0) or (i%2 == 1 and j%2 == 1):
                    self.grid[i,j]= [1,1,1]

    def render(self):
        plt.imshow(self.grid)

    def add_red(self,x,y):
        lst = []
        lst.append(x)
        lst.append(y)
        ChessBoard.red.append(lst)
        self.grid[x,y] = [1,0,0]

    def add_blue(self,x,y):
        lst = []
        lst.append(x)
        lst.append(y)
        ChessBoard.blue.append(lst)
        self.grid[x,y] = [0,0,1]

    @staticmethod
    def is_diagonal(ax,ay,bx,by):
        """
        given coordinates for 2 points a and b
        determine if they are diagonals
        """
        dx = abs(ax - bx)
        dy = abs(ay - by)
        return dx == dy
    
    def is_under_attack(self):
        check = False

        result = ChessBoard.is_diagonal(ChessBoard.red[0][0],ChessBoard.red[0][1],ChessBoard.blue[0][0]        ,ChessBoard.blue[0][1])
        if result:
            check = True

        if ChessBoard.red[0][0] == ChessBoard.blue[0][0]:
            check = True

        if ChessBoard.red[0][1] == ChessBoard.blue[0][1]:
            check = True

        return check
```

```python
board = ChessBoard()
board.add_red(0,3)
board.add_blue(3,3)
board.render()
print(board.is_under_attack())
#check Vertical
assert board.is_under_attack() == True
print('***** Test One Pass  *****')
```

```python
board2 = ChessBoard()
board2.add_red(3,5)
board2.add_blue(3,3)
board2.render()
#check horizental
assert board.is_under_attack() == True
print('***** Test Tow Pass  *****')
```

```python
board3 = ChessBoard()
board3.add_red(0,0)
board3.add_blue(7,7)
board3.render()
#check diagonal
assert board.is_under_attack() == True
print('***** Test Three Pass  *****')
```

```python
board4 = ChessBoard()
board4.add_red(7,0)
board4.add_blue(3,5)
board4.render()
#check No Attack
assert board.is_under_attack() == False
print('***** Test Four Pass  *****')
```
